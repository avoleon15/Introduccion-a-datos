## **EJERCICIO 2**

**Crear un diagrama de dispersión, scatter plot**

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["degree_days_heating"], df["humidity_relative_avg"])
plt.title("Relación entre Heating Degree Days y Humedad Promedio")
plt.xlabel("Heating Degree Days")
plt.ylabel("Humedad Relativa Promedio")
plt.show()
```

```python
Muestra la relación entre los heating degree days y la humedad promedio. Permite observar si existe alguna relación entre el clima frío y los niveles de humedad.
```

**Crear un gráfico de barras (bar plot)**

```python
humedad_pais = df.groupby("country_code")["humidity_relative_avg"].mean()

plt.figure()
plt.bar(humedad_pais.index, humedad_pais.values)
plt.title("Humedad promedio por país")
plt.xlabel("País")
plt.ylabel("Humedad promedio")

plt.xticks(rotation=45)
plt.show()
```

Compara la humedad promedio entre diferentes países. Sirve para identificar qué países tienen mayor o menor nivel de humedad.

**Crear un histograma (histogram)**

```python
plt.figure()
plt.hist(df["humidity_relative_avg"], bins=20)
plt.title("Distribución de la humedad relativa promedio")
plt.xlabel("Humedad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
Muestra cómo se distribuyen los valores de humedad en el dataset. Permite ver si la mayoría de los datos se concentran en ciertos rangos de humedad.
```

**¿Cómo se podría identificar el tipo de gráfico ideal para presentar información de un dataset?**

- El tipo de gráfico ideal depende del tipo de datos y del objetivo del análisis: si se quiere analizar la relación entre variables se usa un diagrama de dispersión, si se busca comparar categorías se utiliza un gráfico de barras, y si se desea observar la distribución de los datos se usa un histograma.
