#####¿En qué parte o partes de un Pipeline de Datos se pueden usar las operaciones de Lectura (R) y Escritura (W)?

##### -Respuesta: Write se pueda hacer a la hora de editar/filtrar los datos mientras que Read se usa para leer y analizar los datos

````python
import pandas as pd
import csv
````

````python
with open("LBtext.txt", "r") as f:
    contenido = f.read()

print(contenido)
````

````python
with open("LBcsv.csv", "r") as f:
    df = pd.read_csv(f)

print(df)
````

````python
with open("LBcsv.csv", "w", newline="") as f:
    writer = csv.writer(f)

    writer.writerow(["Nombre", "Apellido","Edad"])
    writer.writerow(["Alvaro", "Leon", 21])
    writer.writerow(["Valentina", "Moreno", 30])
    writer.writerow(["Oliver", "Leon", 5])
    writer.writerow(["Vanny", "Moreno", 12])
````

````python
with open("LBtext.txt", "w") as f:
    f.write("Hola mundo\n")
    f.write("Alvaro Leon\n")
    f.write("Valentina Moreno\n")
    f.write("Oliver Leon\n")
    f.write("Vanny Moreno\n")
````

#EJERCICIO 3

####Leer un archivo Pickle
####Escribir un archivo Pickle


#####¿Qué tipo de dato se podría almacenar con un Pickle?
#####-Respuesta: Se puede almacenar cualquier tipo de datos

````python
import pickle
````

````python
data = {
    "nombre": "Alvaro",
    "apellido": "Leon",
    "edad": 21
}

with open("datos.pkl", "wb") as f:
    pickle.dump(data, f)
````

````python
with open("datos.pkl", "rb") as f:
    data = pickle.load(f)

print(data)
````
